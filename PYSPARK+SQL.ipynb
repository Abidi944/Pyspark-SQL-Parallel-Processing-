```python
from pyspark.sql import SparkSession
import psutil
```

```python
spark = SparkSession.builder.appName("Indonesian Economical Data 2021").getOrCreate()
```

```python
economic_df = spark.read.csv("/content/sample_data/2021socio_economic_indonesia.csv", header=True)
print(economic_df.head())
```

```python
economic_df.show()
```

```python
economic_df.summary().toPandas()
```

```python
economic_df.createOrReplaceTempView("EcoData21")
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE poorpeople_percentage >=40 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE poorpeople_percentage <=2.6 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE reg_gdp >=700.000 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE reg_gdp <=1.200 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE life_exp >=77 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE life_exp <=57 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE avg_schooltime >=12 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE avg_schooltime <3 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE exp_percap >= 20000 ORDER BY province ASC")
result.toPandas()
```

```python
result = spark.sql("SELECT * FROM EcoData21 WHERE exp_percap <= 4500 ORDER BY province ASC")
result.toPandas()
```
